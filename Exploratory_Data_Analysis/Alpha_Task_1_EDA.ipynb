# **Alpha-task 1. Exploratory Data Analysis**

**Введение**

Итак, неотъемлемая часть работы аналитика, а значит, твоей работы, – это проверка гипотез. Для этого тебе придётся работать с данными из самых разных источников, и далеко не всегда ты будешь знать заранее, что они из себя представляют. Поэтому один из ключевых навыков, которым ты должен владеть, – умение выполнять **разведочный анализ данных (англ. exploratory data analysis, EDA)**.

Выполняя такой анализ, аналитик принимает на вход неизвестные ему с точки зрения внутренней структуры данные (он не знает, сколько в них пропусков, какое распределение они имеют, за какой период собраны и т. д.), преобрабатывает их и отвечает на интересующие его вопросы на их основе.

Разведочный анализ данных можно сравнить с археологическими раскопками в мире информации. Это не просто рутинная предварительная обработка, а захватывающее путешествие, во время которого ты выявляешь неожиданные моменты, скрытые паттерны и закономерности. Кроме того, EDA служит защитой от неприятных сюрпризов. Ты можешь найти аномалии в данных, которые искажают выводы, и принять меры для их исправления, определить недостающие данные и заполнить пробелы.

**Шаги разведочного анализа данных:**

**1. Понимание структуры и характеристик набора данных:**
На этом этапе ты внимательно изучаешь структуру и характеристики данных, чтобы понять, с чем работаешь. Это включает в себя анализ размера набора данных, типов переменных, наличия пропущенных значений и дубликатов. Такой подход формирует четкое представление о данных и становится основой для дальнейшего анализа.

**2. Выявление аномалий и выбросов:**
Данные могут содержать сюрпризы, влияющие на выводы. С помощью методов EDA ты можешь обнаружить аномалии и выбросы, которые могут возникать из-за ошибок ввода или случайных событий. Определение и устранение таких аномалий критически важно для обеспечения надежности результатов.

**3. Идентификация связей и корреляций между переменными:**
Статистические методы EDA помогают выявить взаимосвязи между переменными, что позволяет понять влияние одних факторов на другие. Это может стать основой для формулирования гипотез и принятия бизнес-решений.

**4. Подготовка данных для дальнейших этапов анализа:**
На этом этапе ты обрабатываешь данные для подготовки их к более сложным аналитическим методам. Очищаешь данные от шума, заполняешь пропуски и проводишь масштабирование или преобразования переменных, обеспечивая качественную структуру для точных результатов в дальнейшем анализе.

##  **5. Анализ числовых полей**

```python
def optim_bins(
    series: pd.Series,
    show_size: bool = False
    ) -> int:
    """
    Расчёт количества бинов для построения гистограмм
    :param series: колонка данных
    :param show_size: параметр демонстрации размера одного бина (по умолчанию не показывает)
    :return: количество бинов
    """
    BINS = 1 + m.ceil(m.log(len(series), 2))
    BINS_SIZE = len(series) / BINS

    if show_size:
        return BINS, BINS_SIZE
    else:
        return BINS
```

```python
features = ['loan_amnt', 'funded_amnt', 'int_rate', 'installment', 'sub_grade', 'home_ownership']
features_ru = [feature_names_dict[f] for f in features]
n_features = len(features)

fig, axes = plt.subplots(nrows=n_features, ncols=1, figsize=(10, 4 * n_features))

# Если одна переменная, axes не будет массивом
if n_features == 1:
    axes = [axes]

for i_feature, (feature, feature_ru) in enumerate(zip(features, features_ru)):
    ax = axes[i_feature]
    plot_data = data[feature].sort_values()
    sns.histplot(
        x=plot_data,
        kde=False,
        bins=optim_bins(data[feature]),
        ax=ax
    )
    ax.set_title(feature_ru)
    ax.set_xlabel('Значение')
    ax.set_ylabel('Количество записей')

plt.tight_layout()
plt.show()
```
