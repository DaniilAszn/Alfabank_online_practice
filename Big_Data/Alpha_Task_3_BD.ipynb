**Как организована инфраструктура Больших данных в Альфа-Банке**

В банковской сфере собираются и анализируются огромные объемы информации, например — данные о карточных трансакциях, миллиарды которых совершаются ежедневно. Эти данные можно использовать для улучшения клиентского опыта или для допродажи продуктов и услуг действующим клиентам. Однако для эффективного использования таких данных необходима специальная инфраструктура. В Альфа-Банке в эту инфраструктуру входят:
1. **Платформа Big Data**, аккумулирующая в себе максимально полный набор данных, получаемых из различных источников банка, а также внешних источников (все вместе — **Datalake**), и предназначенная для решения задач Ad-hoc, доступа к данным для аналитиков банка, для проверки гипотез, построения пилотных решений и построения ML-моделей. **Для хранения данных** используется **кластер Apache Hadoop**.
2. **Для обработки больших объемов данных** используется **кластер Apache Spark**, который позволяет выполнять распределенные параллельные вычисления и обрабатывать данные в реальном времени.
3. Платформа **для разработки**. На ней представлено **все необходимое ПО для разработки и автоматизации**.

Основу этой инфраструктуры — пункты 1 и 2 — поддерживает **команда BI / DWH**, которая предоставляет инструменты, процессы, документацию и консультирование для работы с данными, разрабатывает и поддерживает кластеры Apache Spark, Apache Hadoop и т.д. Ты уже знаешь об этой команде из 4-го дня практики, но давай повторим, чтобы лучше закрепилось:
* **BI (Business intelligence, бизнес-аналитика)** — обозначение компьютерных методов и инструментов для организаций, обеспечивающих перевод транзакционной деловой информации в читаемую для человека форму, а также средства для массовой работы с такой обработанной информацией.
* **DWH (Data Warehouse, хранилище данных)** — это специализированная система для хранения и управления большими объемами данных, которые объединяются из разных источников с целью анализа и построения отчетов.

Пункт 3 находится в зоне ответственности также знакомой тебе команды MLOps. Команда поддерживает платформу разработки, хранения кода, внедряет принципы DevOps в сферу машинного обучения:
* **MLOps (Machine Learning Operations)** — практический подход к разработке, развертыванию и управлению моделями машинного обучения (ML) в продукционной среде.

````python
import os
import zipfile
import pandas as pd
import seaborn as sns
import dask.dataframe as dd
import dask_ml.preprocessing
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from IPython.display import display, HTML
display(HTML("<style>.container {width:90% !important; }</style>"))

import warnings
warnings.filterwarnings("ignore")
````

## **1. Загрузка данных и инициализация dask.dataframe**

````python
# Распаковываем архив

%%time
zip_path = '/content/drive/MyDrive/data/cardtransactions.zip'
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall('.')
````

````python
# Загружаем данные из всех партиций

data_path = 'card_transactions'
df = dd.read_parquet(os.path.join(data_path, '*.parquet'))
````

````python
%%time
df.shape[0].compute()
````

## **2. Формирование признаков**

````python
%%time

# def mode_function(series):
#     return series.mode()[0] if not series.mode().empty else None


# 1 Средняя сумма транзакции
mean_amount = (
    df.groupby('client_id')['amnt']
    .mean()
    .rename('mean_amount')
    .repartition(npartitions=1)
    .reset_index(drop=True)
)

# 2 Общая сумма транзакций
total_amount = (
    df.groupby('client_id')['amnt']
    .sum()
    .rename('total_amount')
    .repartition(npartitions=1)
    .reset_index(drop=True)
)

# 3 Количество транзакций
transaction_count = (
    df.groupby('client_id')
    .size()
    .rename('transaction_count')
    .repartition(npartitions=1)
    .reset_index(drop=True)
)

# 4 Количество используемых валют
currency_diversity = (
    df.drop_duplicates(['client_id', 'currency'])
    .groupby('client_id')['currency']
    .count()
    .rename('currency_diversity')
    .repartition(npartitions=1)
    .reset_index(drop=True)
)

# 5 Количество используемых операций
operation_kind_diversity = (
    df.drop_duplicates(['client_id', 'operation_kind'])
    .groupby('client_id')['operation_kind']
    .count()
    .rename('operation_kind_diversity')
    .repartition(npartitions=1)
    .reset_index(drop=True)
)

# 6 Количество используемых типов карт
card_type_diversity = (
    df.drop_duplicates(['client_id', 'card_type'])
    .groupby('client_id')['card_type']
    .count()
    .rename('card_type_diversity')
    .repartition(npartitions=1)
    .reset_index(drop=True)
)

# 7 Количество используемых групп операций по картам
card_group_diversity = (
    df.drop_duplicates(['client_id', 'operation_type_group'])
    .groupby('client_id')['operation_type_group']
    .count()
    .rename('card_group_diversity')
    .repartition(npartitions=1)
    .reset_index(drop=True)
)

# 8 Доля транзакций с использованием электронной коммерции
ecommerce_ratio = (
    (df['ecommerce_flag'] == 1)
    .groupby(df['client_id'])
    .mean()
    .rename('ecommerce_ratio')
    .repartition(npartitions=1)
    .reset_index(drop=True)
)

# 9 Количество используемых платёжных систем
payment_system_diversity = (
    df.drop_duplicates(['client_id', 'payment_system'])
    .groupby('client_id')['payment_system']
    .count()
    .rename('payment_system_diversity')
    .repartition(npartitions=1)
    .reset_index(drop=True)
)

# 10 Доля транзакций с внесением средств
deposit_ratio = (
    (df['income_flag'] == 1)
    .groupby(df['client_id'])
    .mean()
    .rename('deposit_ratio')
    .repartition(npartitions=1)
    .reset_index(drop=True)
)

# 11 Количество уникальных идентификаторов типа торговой точки
mcc_diversity = (
    df.drop_duplicates(['client_id', 'mcc'])
    .groupby('client_id')['mcc']
    .count()
    .rename('mcc_diversity')
    .repartition(npartitions=1)
    .reset_index(drop=True)
)

# 12 Количество стран, где клиент совершал платежи
country_diversity = (
    df.drop_duplicates(['client_id', 'country'])
    .groupby('client_id')['country']
    .count()
    .rename('country_diversity')
    .repartition(npartitions=1)
    .reset_index(drop=True)
)

# 13 Количество городов, где клиент совершал платежи
city_diversity = (
    df.drop_duplicates(['client_id', 'city'])
    .groupby('client_id')['city']
    .count()
    .rename('city_diversity')
    .repartition(npartitions=1)
    .reset_index(drop=True)
)

# 14 Среднее количество часов между транзакциями
hour_diff_mean = (
    df.groupby('client_id')['hour_diff']
    .mean()
    .rename('hour_diff_mean')
    .repartition(npartitions=1)
    .reset_index(drop=True)
)

# 15 Среднее количество транзакций в неделю
transactions_per_week_mean = (
    df.groupby(['client_id', 'weekofyear'])['amnt']
    .count()
    .groupby('client_id')
    .mean()
    .rename('transactions_per_week_mean')
    .repartition(npartitions=1)
    .reset_index(drop=True)
)

# 16 Доля транзакций в выходные дни
weekend_ratio = (
    (df['day_of_week'] >= 5)
    .groupby(df['client_id'])
    .mean()
    .rename('weekend_ratio')
    .repartition(npartitions=1)
    .reset_index(drop=True)
)
````

````python
%%time

features_df = dd.concat([
    mean_amount,
    total_amount,
    transaction_count,
    currency_diversity,
    operation_kind_diversity,
    card_type_diversity,
    card_group_diversity,
    ecommerce_ratio,
    payment_system_diversity,
    deposit_ratio,
    mcc_diversity,
    country_diversity,
    city_diversity,
    hour_diff_mean,
    transactions_per_week_mean,
    weekend_ratio
], axis=1)

features_df = features_df.compute()
````

## **3. Масштабирование признаков**

````python
scaler = StandardScaler()
features_df_scaled = pd.DataFrame(scaler.fit_transform(features_df), columns=features_df.columns)
````

## **4. Подбор оптимального количества кластеров методом локтя**

````python
# Определим диапазон возможного количества кластеров

range_n_clusters = list(range(1, 11))
````

````python
# Вычислим ошибку для каждого количества кластеров

inertia = []
for n_clusters in tqdm(range_n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, init='k-means++')
    kmeans.fit(features_df_scaled)
    inertia.append(kmeans.inertia_)
````

````python
# Построим график зависимости ошибки от количества кластеров

plt.figure(figsize=(10, 6))
plt.plot(range_n_clusters, inertia, marker='o')
plt.title('Метод локтя')
plt.xlabel('Количество кластеров')
plt.ylabel('Ошибка (Inertia)')
plt.grid(True)
plt.show()
````

## **5. Обучение модели K-Means для выбранного количества кластеров**

````python
# Обучение модели K-Means

optimal_n_clusters = 5
kmeans = KMeans(n_clusters=optimal_n_clusters, random_state=42, init='k-means++')
kmeans.fit(features_df_scaled)
````

````python
# Добавление меток кластеров в исходный датафрейм

features_df['cluster'] = kmeans.labels_ + 1
feature_cols = list(features_df.columns)
feature_cols.remove('cluster')
````

````python
features_df['cluster'].value_counts(normalize=True)
````

## **6. Анализ полученных результатов**

````python
discrete_features = [
    'operation_kind_diversity',
    'card_type_diversity',
    'card_group_diversity',
    'payment_system_diversity'
]

features_ru_names = {
    'mean_amount': 'Средняя сумма транзакции',
    'total_amount': 'Общая сумма транзакций',
    'transaction_count': 'Количество транзакций',
    'currency_diversity': 'Количество используемых валют',
    'operation_kind_diversity': 'Количество используемых типов транзакций',
    'card_type_diversity': 'Количество используемых типов карт',
    'card_group_diversity': 'Количество используемых групп операций по картам',
    'ecommerce_ratio': 'Доля транзакций с использованием электронной коммерции',
    'payment_system_diversity': 'Количество используемых платёжных систем',
    'deposit_ratio': 'Доля транзакций с внесением средств',
    'mcc_diversity': 'Количество уникальных идентификаторов типа торговой точки',
    'country_diversity': 'Количество стран, где клиент совершал платежи',
    'city_diversity': 'Количество городов, где клиент совершал платежи',
    'hour_diff_mean': 'Среднее количество часов между транзакциями',
    'transactions_per_week_mean': 'Среднее количество транзакций в неделю',
    'weekend_ratio': 'Доля транзакций в выходные дни'
}
````

````python
n_cols = 2
n_rows = (len(feature_cols) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 5))
axes = axes.flatten()

for i, feature_col in enumerate(feature_cols):
    ax = axes[i]

    if feature_col in discrete_features:
        shrink = 1 if features_df[feature_col].nunique() < 5 else 10

        sns.histplot(
            x=features_df[feature_col],
            hue=features_df['cluster'],
            palette='Set1',
            multiple='dodge',
            shrink=shrink,
            stat='percent',
            common_norm=False,
            ax=ax
        )
        ax.set_title(f'{features_ru_names[feature_col]}')
        ax.set_xlabel('Значение признака')
        ax.set_ylabel('Процент')
    else:
        sns.kdeplot(
            data=features_df,
            x=feature_col,
            hue='cluster',
            palette='Set1',
            fill=True,
            common_norm=False,
            alpha=0.25,
            ax=ax
        )
        ax.set_title(f'{features_ru_names[feature_col]}')
        ax.set_xlabel('Значение признака')
        ax.set_ylabel('Плотность вероятности')

plt.tight_layout()
plt.show()
````
