**В качестве ключевых задач ML в аналитике данных можно назвать:**
* **Выявление скрытых закономерностей, тенденций и взаимосвязей** в данных, которые могут быть слишком сложны для обнаружения традиционными методами.

**Если сузить это именно до банковской сферы, то вот несколько ключевых областей, где применяется ML:**
1. **Оценка кредитоспособности:** банки используют алгоритмы машинного обучения для анализа кредитных заявок и оценки рисков. Модели могут учитывать множество факторов, таких как кредитная история, доход, занятость и даже поведенческие данные, чтобы предсказать вероятность дефолта заёмщика.
2. **Обнаружение мошенничества:** ML помогает в выявлении аномалий в транзакциях, что позволяет быстро реагировать на возможные случаи мошенничества. Алгоритмы могут обучаться на исторических данных о мошеннических транзакциях и выявлять подозрительные паттерны в реальном времени.
3. **Персонализированные предложения:** с помощью анализа данных о клиентах банки могут создавать персонализированные предложения и рекомендации. ML-алгоритмы помогают сегментировать клиентов по различным параметрам и предлагать им продукты, которые наилучшим образом соответствуют их потребностям.
4. **Управление рисками:** машинное обучение позволяет банкам более точно оценивать риски, связанные с инвестициями и кредитованием. Модели могут анализировать рыночные данные и предсказывать изменения, что помогает в принятии более обоснованных решений.
5. **Оптимизация операций:** ML может использоваться для оптимизации внутренних процессов банка, таких как обработка заявок, управление активами и прогнозирование спроса на различные финансовые продукты.
6. **Анализ клиентских данных:** банки могут использовать ML для анализа поведения клиентов, выявления трендов и предпочтений, что позволяет улучшать качество обслуживания и разрабатывать новые продукты.
7. **Обслуживание клиентов:** чат-боты и виртуальные ассистенты, основанные на ML, помогают клиентам получать ответы на вопросы и решать проблемы 24/7, улучшая клиентский опыт.

**Как аналитики взаимодействуют с машинным обучением в своей работе в Альфа-Банке?**
1. **Анализируют результаты работы ML-модели** (оценку, формируемую моделью, её предсказания). Например, риск-аналитики используют прогнозы PD-моделей, чтобы анализировать портфель.

**Задача классификации**

**Бинарные классификаторы** являются, пожалуй, наиболее популярными методами машинного обучения, применяемыми на практике. В банковской сфере модели бинарной классификации используются практически во всех ключевых направлениях деятельности:
* в скоринге — это модели прогнозирования вероятности дефолта заёмщика;
* в перекрестных продажах (CRM) — модели отклика клиента на x-sell предложение;
* во взыскании — модели прогнозирования вероятности возврата просроченного кредита;
* в антифроде — модели выявления мошеннических кредитов;
* и т.д.

В бинарной классификации объекты выборки делятся на два класса — положительные (positive) и отрицательные (negative). Например, применительно к банковским задачам это могут быть следующие разбиения:
* попадет кредит в просрочку (1 — positive) или нет (0 — negative);
* откликнется клиент на x-sell предложение банка (1) или нет (0);
* вернёт клиент просроченный кредит (1) или не вернёт (0);
* окажется кредитная заявка мошеннической (1) или нет (0).

Соответственно, модель бинарной классификации присваивает объектам одну из двух меток — positive или negative. А поскольку моделям свойственно ошибаться, то в результате бинарной классификации все объекты выборки разбиваются на четыре типа, образуя матрицу ошибок:
1. истинно положительные (true positive — TP);
2. истинно отрицательные (true negative — TN);
3. ложноположительные (false positive — FP);
4. ложноотрицательные (false negative — FN).

**Метрики качества**

Метрики качества бинарных классификаторов делятся на «пороговые» (single-threshold) и «не зависящие от порога» (threshold-free).

Бинарные классификаторы обычно прогнозируют вероятность принадлежности объекта к одному из двух классов. То есть каждому объекту присваивается числовое значение в диапазоне от 0 до 1. Таким образом, чтобы разделить прогнозные значения на два класса, необходимо выбрать порог, разбивающий весь диапазон значений на две группы. Примеры популярных пороговых метрик: Accuracy, Recall, Precision, F1.

Если переменные в матрице ошибок рассчитать для различных порогов, то можно построить метрики качества, не зависящие от порога, такие как AUC ROC, Gini и др.

**Примеры метрик для задач классификации**
* **Accuracy:**
    * Описание: Доля правильно классифицированных экземпляров от общего числа экземпляров
    * Формула: Accuracy = (TP + TN) / (TP + TN + FP + FN)
    * Плюсы: Простота интерпретации
    * Минусы: Может быть неинформативной в случае несбалансированных классов
* **Recall:**
    * Описание: Доля истинно положительных результатов среди всех реальных положительных случаев
    * Формула: Recall = TP / (TP + FN)
    * Плюсы: Полезна, когда важно минимизировать количество ложноотрицательных результатов
    * Минусы: Не учитывает ложноположительные результаты
* **Precision:**
    * Описание: Доля истинно положительных результатов среди всех положительных предсказаний
    * Формула: Precision = TP / (TP + FP)
    * Плюсы: Полезна, когда важно минимизировать количество ложноположительных результатов
    * Минусы: Не учитывает ложноотрицательные результаты
* **F1:**
    * Описание: Гармоническое среднее между Precision и Recall. Используется для оценки модели, когда требуется баланс между точностью и полнотой
    * Формула: F1 = 2 ⋅ Precision ⋅ Recall / (Precision + Recall)
    * Плюсы: Учитывает как точность, так и полноту, что делает его полезным для несбалансированных классов
    * Минусы: Не даёт информации о количестве истинно отрицательных результатов
* **ROC AUC и Gini:**
    * Описание: Показывает соотношение между истинно положительными и ложноположительными значениями при различных порогах классификации. Чем ближе AUC или Gini к 1, тем лучше модель
    * Формула: AUC = площадь под ROC-кривой; Gini = 2 ⋅ AUC - 1
    * Плюсы: Простота интерпретации. Полезный показатель для сравнения производительности двух разных моделей
    * Минусы: Может быть неинформативной в случае несбалансированных классов

    * оценить качество ранжирования финального скора по системе светофора (усреднив значения Gini всех месячных срезов):
        * значение Gini меньше 35 п.п. — красный сигнал,
        * между 35 и 50 — жёлтый,
        * больше 50 — зелёный.

````python
from typing import *
import json
from zipfile import ZipFile
import numpy as np
import pandas as pd
import plotly.io as pio
from datetime import datetime
from dateutil.relativedelta import relativedelta
from sklearn.metrics import roc_auc_score
from scipy.stats import norm
import plotly.graph_objs as go
from plotly.subplots import make_subplots

import warnings
warnings.filterwarnings("ignore")
````

````python
def load_json_data_from_zip(
    zip_path: str
    ) -> pd.DataFrame:
    """
    Парсинг структуры zip-файла с json'ами
    :param zip_path: путь до зип файла
    :return: сформированный датафрейм
    """
    with ZipFile(zip_path, 'r') as z:
        json_files = [f for f in z.namelist() if f.startswith('jsons/') and f.endswith('.json')]
        data = [
            {
                'loanID': record['loanID'],
                'eventTime': datetime.strptime(record['eventTime'], '%Y-%m-%d %H:%M:%S.%f'),
                'final_model_score': record['values']['final_model_score'],
                'internal_model_score': record['values']['internal_model_score'],
                'external_model_score': record['values']['external_model_score']
            }
            for file in json_files
            for record in [json.load(z.open(file))]
            if record['modelID'] == 'pd-auto-online'
        ]

    return pd.DataFrame(data)


def gini(
    y_true: Union[pd.Series, np.array],
    y_scores: Union[pd.Series, np.array]
    ) -> float:
    """
    Расчёт значения коэффициента Gini
    :param y_true: метки классов
    :param y_scores: скоры модели
    :return: значение Gini
    """
    return 2 * roc_auc_score(y_true, y_scores) - 1


def confidence_intervals(
    data: List,
    confidence_level: float = 0.95
    ) -> tuple:
    """
    Построение доверительного интервала
    :param data: выборка данных
    :param confidence_level: уровень доверия (по умолчанию равен 0.95)
    :return: границы доверительного интервала
    """
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(len(data))
    margin_of_error = norm.ppf((1 + confidence_level) / 2) * std_err

    return mean - margin_of_error, mean + margin_of_error


def bootstrap_gini_ci(
    y_true: Union[pd.Series, np.array],
    y_scores: Union[pd.Series, np.array],
    n_bootstraps: int = 1000,
    confidence_level: float = 0.95
    ) -> tuple:
    """
    Построение доверительного интервала для значений Gini
    :param y_true: метки классов
    :param y_scores: скоры модели
    :param n_bootstraps: количество итераций бутстрепа (по умолчанию равно 1000)
    :param confidence_level: уровень доверия (по умолчанию равен 0.95)
    :return: границы доверительного интервала для значений Gini
    """
    gini_values = [
        gini(y_true.sample(frac=0.7, replace=True), y_scores.sample(frac=0.7, replace=True))
        for _ in range(n_bootstraps)
    ]

    return confidence_intervals(gini_values, confidence_level)


def rank_quality(
    gini_value: float
    ) -> str:
    """
    Определение сигнала коэффициента Gini по системе светофора
    :param gini_value: значение коэффициента Gini
    :return: цвет сигнала
    """
    if gini_value < 0.35:
        return 'Красный сигнал'
    elif 0.35 <= gini_value <= 0.5:
        return 'Жёлтый сигнал'
    else:
        return 'Зелёный сигнал'
````

## **1. Загрузка данных и формирование выборки для проведения аналитики**

````python
zip_file_path = 'jsons.zip'
delq_file_path = 'deal_delqs_logs.pq'

json_df = load_json_data_from_zip(zip_file_path)
delq_df = pd.read_parquet(delq_file_path)
````

````python
# Приводим типы

delq_df['period_start_date'] = pd.to_datetime(delq_df['period_start_date'])
delq_df['period_end_date'] = pd.to_datetime(delq_df['period_end_date'])
delq_df['delq_days'] = (delq_df['period_end_date'] - delq_df['period_start_date']).dt.days
json_df['threshold_date'] = json_df['eventTime'] + pd.DateOffset(months=4)
````

````python
# Фильтруем пустые скоры

json_df = json_df[json_df['final_model_score'].notnull()]
````

````python
# Находим сделки с нужной просрочкой в нужном интервале

delq_df_joined = delq_df.merge(
    json_df[['loanID', 'eventTime', 'threshold_date']],
    how='left', left_on='loan_id', right_on='loanID'
).query("(period_start_date >= eventTime) & (period_start_date <= threshold_date) & (delq_days >= 30)")

default_loan_id = delq_df_joined['loan_id'].unique()
json_df['default_flag'] = json_df['loanID'].isin(default_loan_id).astype(int)
````

## **2. Расчёты**

````python
# Считаем Джини

json_df['event_month'] = json_df['eventTime'].dt.to_period('M')

gini_results = json_df.groupby('event_month').apply(
    lambda x: pd.Series({
        'gini_final': gini(x['default_flag'], x['final_model_score']),
        'gini_internal': gini(x['default_flag'], x['internal_model_score']),
        'gini_external': gini(x['default_flag'], x['external_model_score'])
    })
).reset_index()
````

````python
# Считаем доверительные интервалы

ci_results = json_df.groupby('event_month').apply(
    lambda x: pd.Series({
        'ci_lower': bootstrap_gini_ci(x['default_flag'], x['internal_model_score'])[0],
        'ci_upper': bootstrap_gini_ci(x['default_flag'], x['internal_model_score'])[1]
    })
).reset_index()
````

````python
# Приводим типы

gini_results['event_month'] = gini_results['event_month'].astype(str)
gini_results['upift'] = gini_results['gini_final'] - gini_results['gini_internal']
ci_results['event_month'] = ci_results['event_month'].astype(str)
ci_results['quality'] = ci_results['ci_lower'].apply(rank_quality)
````

## **3. Выводы**

````python
fig = make_subplots()

for col, name in zip(['gini_final', 'gini_internal', 'gini_external'],
                     ['Джини финального скора', 'Джини внутреннего скора', 'Джини внешнего скора']):
    fig.add_trace(
        go.Scatter(
            x=gini_results['event_month'],
            y=gini_results[col],
            mode='lines+markers+text',
            name=name,
            text=gini_results[col] * 100
        )
    )

fig.update_layout(title='Ранжирование')
fig.update_yaxes(tickformat='.0%')
fig.update_traces(texttemplate='%{text:.2f}', textposition='top center')
fig.show()
````
